## Part II: Cournot Duopoly


Cournot Duopoly is a classic static game that models the imperfect competition in which multiple firms (agents) compete in price and production to capture market share. Since the firms' actions are continuous variables, the game is a continuous action setting.
It is a **nonzero-sum game** (neither team-based nor zero-sum) which represents a challenge for current MARL methods.

Let $a_i\in Q_i = [0,A_i]$ represents the set of actions for firm (agent) $i\in\{1,2\ldots, N\}:=\mathcal{N}$,
where $A_i\in \mathbb{R}_{>0}$.
Each firm (agent) $i$'s reward (profit) is 
$$
R_i(a_i,a_{-i})=g_i(a_i,a_{-i}) - w_i(a_i),
$$
where $g_i(a_i,a_{-i})$ is the revenue of firm (agent) $i$ and $w_i(a_i)$ is the cost to firm (agent) $i$ of taking action $a_i$.
In this market scenario, we have $g_i = a_i * p(a)$, where $a=a_1 + \dots + a_N$ and $p$ is the market price of the production and is a decreasing function when it is positive: if the agents' total output increases, then the price decreases (unless it is already zero).

```python
import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from copy import deepcopy
import pandas as pd
#citied to COMP0124 lab1
alpha = 12
beta = 1.0
gamma1 = 1.0
gamma2 = 2.0

def payoff_n_cournot(action_n, i, p):
            """
            Define the payoff function R_i(a_i,a_{-i}).
            :param action_n: (nd.array) a list of all agents' actions, shape is (agent_num,)
            :param i: agent index
            :return: R_i(a_i,a_{-i})
            """
            ########### TODO: Compute R_i(a_i,a_{-i}) (1 point) ###########
            #  implement Eqs. (5)
            if p == 1:
                r = (action_n[i] * (alpha -beta * np.sum(action_n))) - (gamma1 * action_n[i])
            else:
                r = (action_n[i] * (alpha -beta * np.sum(action_n))) - (gamma2 * action_n[i])
            ########### END TODO ############################
            return r


SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title


plt_index = np.round(np.arange(0, 10, 0.1), 3)
X = Y = plt_index


def plot_result(laa, lbb):
    fig, ax = plt.subplots(1, 3, figsize=(15, 4))

    plt_gap = 20
    plt_index = np.round(np.arange(0, 10, 1/plt_gap), 3)
    X, Y = np.meshgrid(plt_index, plt_index)
    ########### TODO: Calculate V^{1} and V^{2} ###########
    def payoff_n_cournot_grid(action_n, i, p):
                """
                Define the payoff function R_i(a_i,a_{-i}).
                :param action_n: (nd.array) a list of all agents' actions, shape is (agent_num,)
                :param i: agent index
                :return: R_i(a_i,a_{-i})
                """
                a_i = action_n[i]
                a_sum = np.zeros(action_n[i].shape)
                for a in action_n:
                    a_sum = np.add(a_sum, a)
                if p == 1:
                    r = a_i * (alpha - beta * a_sum) - gamma1 * a_i
                else: 
                    r = a_i * (alpha - beta * a_sum) - gamma2 * a_i
                return r
    Z_a = payoff_n_cournot_grid(np.array([X, Y]), 0, 1)
    Z_b = payoff_n_cournot_grid(np.array([X, Y]), 1, 2)
    ########### END TODO ################################################

    for Z_, ax_ in zip([Z_a, Z_b], ax):
        df = pd.DataFrame(np.transpose(Z_), index=plt_index, columns=plt_index)
        sns.heatmap(df, ax=ax_)
        ax_.invert_yaxis()
        ax_.set_xlabel("Agent 2's action $a_{2}$")
        ax_.set_ylabel("Agent 1's action $a_{1}$")

    ########### TODO: Calculate the best response curve ###########
    A1_BR_to_A2 = [np.argmax(Z_line) for Z_line in Z_a]
    A2_BR_to_A1 = [np.argmax(Z_line) for Z_line in np.transpose(Z_b)]
    ########### END TODO ################################################
    ax[0].plot(plt_index*plt_gap, A1_BR_to_A2, linewidth=3)
    ax[0].set_title('Utility to Agent 1')
    ax[1].plot(A2_BR_to_A1, plt_index*plt_gap, linewidth=3)
    ax[1].set_title('Utility to Agent 2')
    plt.plot(lbb,laa,'o')
    for i in range(len(lbb)-1):
        plt.annotate('',xy=(lbb[i+1],laa[i+1]),xytext=(lbb[i],laa[i]),arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))

    ax[2].plot(plt_index, np.array(A1_BR_to_A2) / plt_gap, label="A1's BR to A2")
    ax[2].plot(np.array(A2_BR_to_A1) / plt_gap, plt_index, label="A2's BR to A1")
    ax[2].set_title('Best response')
    ax[2].set_xlabel("Agent 2's action $a_{2}$")
    ax[2].set_ylabel("Agent 1's action $a_{1}$")
    ax[2].legend()

    plt.tight_layout()
    plt.show()
```

## Part III: Repeated Game with Markov Stratigies

In a repeated game, an agent can condition its policy on previous history. We now consider a set of simple Markov stratigies where their actions only depend on the results of previous 1 round (i.e., the action of the current time step $t$ only depends on previous time step $t-1$). 

Suppose that two people are involved repeatedly in playing the Prinsoner's Dilemma game, with the payoff of the stage game (the one they play at each time step) as:

$$
\mathbf{R}^1 = \left[\begin{matrix}
2 & 0 \\
3 & 1
\end{matrix}\right] 
\quad 
\mathbf{R}^2 = \left[\begin{matrix}
2 & 3 \\
0 & 1
\end{matrix}\right],
$$

where the action 1 (correpsonds to column/row 1) as "cooperation" (don't confess) and the action 2 (correpsonds to column/row 2) as "defection" (confess).

In the Markov strategy, the policy of the agent conditions on action history of previous 1 round, where both agents' actions are considered. Let's assume that, in the intial step, the action is sampled uniformly from {C, D} for each agent, and then after that each agent's policy is represented (parameterised) by four probabilities (thus mixed stratigies): $(\pi(C|CC),\pi(C|CD),\pi(C|DC),\pi(C|DD)$, C refers to cooperation and D refers to defection). The discount factor is $\delta \in (0,1)$. Thus we have in total eight parameters for the two agents playing against each other.

```python
### Some necessary functions presented in Lab2 notebook
import numpy as np
class iterated_games():
  def __init__(self, payoff1, payoff2, max_step):
    self.payoff1 = payoff1
    self.payoff2 = payoff2
    self.max_step_num = max_step
    self.reset()
  def reset(self):
    self.history = {'agent1':[], 'agent2':[]}
    self.step_num = 0
    return self.history
  def step(self, a1, a2):

    r1, r2 = self.payoff1[a1, a2], self.payoff2[a1, a2]
    self.history['agent1'].append(a1)
    self.history['agent2'].append(a2)
    self.step_num += 1
    done = False
    if self.step_num == self.max_step_num:
      done = True

    return r1, r2, self.history, done

def get_discount_return(return_list, delta):
  return_all = np.array(return_list)
  discount = np.array([delta ** i for i in range(len(return_list))])
  return_all = np.sum(return_all * discount)
  return return_all

def evaluate(env, agent1, agent2, delta):
  history = env.reset()
  done = False
  r1_list = []
  r2_list = []
  while True:
    if done:
      break
    a1 = agent1.step(history)
    a2 = agent2.step(history)
    r1, r2, history, done = env.step(a1, a2)
    r1_list.append(r1)
    r2_list.append(r2)
  return1 = get_discount_return(r1_list, delta) # Discounted return for the first policy
  return2 = get_discount_return(r2_list, delta) # Discounted return for the second policy
  return return1, return2
```
